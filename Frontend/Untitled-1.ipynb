```python
import qrcode
from PIL import Image

def generate_qr(data, filename='qrcode.png', fill='black', back='white'):
    qr = qrcode.QRCode(
        version=1,
        error_correction=qrcode.constants.ERROR_CORRECT_H,
        box_size=10,
        border=4
    )
    qr.add_data(data)
    qr.make(fit=True)
    img = qr.make_image(fill_color=fill, back_color=back)
    img.save(filename)
    print(f'QR Code saved as {filename}')

data = input('Enter text or URL: ')
generate_qr(data)
```
